```python
import numpy as np
import os
import xlrd
import pandas as pd
from sklearn import preprocessing
from matplotlib import pyplot as plt

print(os.getcwd())
swimmers1 = pd.read_csv('./dataset/swimmers5um.csv')
swimmers1
```

```python
for i in range(1,6):
    taj1_X = swimmers1.iloc[:,i*2-1].values
    taj1_Y = swimmers1.iloc[:,i*2].values
    plt.plot(taj1_X[:], taj1_Y[:], '-')
    plt.scatter(taj1_X[:],taj1_Y[:], c=np.arange(taj1_X.size), s = 2)
    plt.colorbar().set_label('Time')
    plt.xlabel('X'); plt.ylabel('Y')
    plt.savefig('./png/5um'+str(i))
    plt.close()
```

```python
i = 1
taj1_X = preprocessing.minmax_scale(swimmers1.iloc[:,i*2-1].values)
taj1_Y = preprocessing.minmax_scale(swimmers1.iloc[:,10].values)
plt.plot(taj1_X[:], taj1_Y[:], '-')
plt.scatter(taj1_X[:],taj1_Y[:], c=np.arange(taj1_X.size), s = 2)
plt.colorbar().set_label('Time')
plt.xlabel('X'); plt.ylabel('Y')
```
